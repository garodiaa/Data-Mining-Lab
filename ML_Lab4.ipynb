```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/`Fall 24/DM_ML/cse.csv',usecols = ['Age','Cabin','Embarked'])
df.head()
```

```python
df.isnull().sum()
```

```python
def input_one(df,var,mean):
  df[var+'_mean'] = df[var].fillna(mean)
```

```python
mean = df['Age'].mean()
print(mean)
```

```python
input_one(df,'Age',mean)
```

```python
df.head()
```

```python
df['Age_mean'].isnull().sum()
```

```python
df['Age'].plot(kind='kde',color='red')
df['Age_mean'].plot(kind='kde',color='green')
```

```python
median = df['Age'].median()
print(median)

mode = df['Age'].mode()
df['Age'].mode()[0] #can pass the direct index also
print(mode)
```

```python
def input_nan(df,variable,mode):
  df[variable+'_mode'] = df[variable].fillna(mode)
```

```python
input_nan(df,'Age',mode[0])
```

```python
df.head()
```

```python
df['Age'].plot(kind='kde',color='red')
df['Age_mean'].plot(kind='kde',color='green')
df['Age_mode'].plot(kind='kde',color='blue')
```

```python
def input_three(df,var,median):
  df[var+'_median'] = df[var].fillna(median)

input_three(df,'Age',median)
```

```python
df['Age'].plot(kind='kde',color='red')
df['Age_mean'].plot(kind='kde',color='green')
df['Age_mode'].plot(kind='kde',color='blue')
df['Age_median'].plot(kind='kde',color='black')
```
