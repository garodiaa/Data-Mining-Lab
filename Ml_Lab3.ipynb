```python
#b
def prime_check(x):
  if x > 1:
    for i in range(2, int(x/2)+1):
      if (x % i) == 0:
        return False
    else:
      return True
  else:
    return False

x = int(input())

if prime_check(x):
  print(x, "is a prime number")
else:
    print(x, "is not a prime number")
```

```python
#c
def fact(x):
  if x==1:
    return 1

  return fact(x-1)*x

x = int(input())
print(f"Factorial of {x} is {fact(x)}")
```

```python
import datetime as dt
print("=====Chatbot=====")
print(f"I am the chatbot, you can ask me the following question!")
ques = ['Hello ChatAi','How are you?','Help me!','Todays Date?','What time it is?']
for q in ques:
  print(f"Q: {q}")

while(True):
  prompt = input("Your Prompt : ")
  if(prompt == ques[0]):
    print(f"Hello there, I am ChatAi")
  elif(prompt == ques[1]):
    print(f"I am fine, thank you for asking")
  elif(prompt == ques[2]):
    print(f'Here are my compatible commands')
    for q in ques:
      print(f"Q: {q}")
  elif(prompt == ques[3]):
    print(f"Today is {dt.datetime.now().date()}")
  elif(prompt == ques[4]):
    print(f"It is {dt.datetime.now().time()}")

  if(prompt == 'quit'):
    print('Signing Out! Thank You')
    break
```

```python
#c
print("Sourav's Calculator")
operations = ['Add','Subtract','Multiply','Divide']
while(True):
  op = input("Operation : ")
  if(op == operations[0]):
    print(f"Input x + y")
    x,y = map(int,input().split())
    print(f"{x} + {y} = {x+y}")
  elif(op == operations[1]):
    print(f"Input x - y")
    x,y = map(int,input().split())
    print(f"{x} - {y} = {x-y}")
  elif(op == operations[2]):
    print(f"Input x x y")
    x,y = map(int,input().split())
    print(f"{x} - {y} = {x*y}")
  elif(op == operations[3]):
    print(f"Input x / y")
    x,y = map(int,input().split())
    print(f"{x} / {y} = {x/y}")
  elif(op == 'quit'):
    print('Signing Out! Thank You')
    break
```
