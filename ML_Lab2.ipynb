```python
df = pd.read_csv('/content/cse.csv',usecols=(['Age','Cabin','Embarked']))
df
```

```python
mean = df['Age'].mean()
mean
```

```python
def input_nan(df,variable,mean):
  df[variable+'_mean'] = df[variable].fillna(mean)
```

```python
input_nan(df,'Age',mean)
```

```python
df
```

```python
df['Age_mean'].isnull().sum()
```

```python
median = df['Age'].median()
print(median)

mode = df['Age'].mode()
print(mode)
```

```python
def input_nan(df,variable,median):
  df[variable+'_median'] = df[variable].fillna(median)
```

```python
input_nan(df,'Age',median)
```
